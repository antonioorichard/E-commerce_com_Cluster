Modelo RFM - RFM Segmentation


Intepretação

    * Hibernating    ( clinte que n compra mais ou entrou em churn)
    * About to Sleep ( Quase em churn)
    * Promising      ( Cliente promissor que fez uma compra regularmente)
    * New Customers  ( Cliente novo que fez uma compra na última semana)
        I. Campanha de Marketing para conhecer o produto.
    * At Risk        ( Um cliente que compra em grande quantidade, mas não está comprando mais, em risco)
    * Need Attention ( cliente está comprando menos em tempo cada vez maior, precisa de atenção)
    * Potential Loyalists (Cliente que tem um potêncial de ser bom cliente, pois compra pouca quantidade regularmente, deve trabalhar neles para comprar mais.)
        I. Fazer eles comprar mais.
    * About to Loose ( Cliente que compra em grandes quantidade, mas faz muito tempo que não compra, eles são valiosos.)
        I. campanha de marketing para trazer eles de volta. 
    * Loya Customers ( Leal cliente que compra em boas quantidade.) 
       I. Tenho que fazer virar Champions
    * Champions      ( Nossa meta de clientes, deve buscar clientes iguais a esse. )
       I. Tenho que manter esses clientes aqui.
       II. Entender esses clientes e buscar iguais a esse

OBS: Esse RFM é bom para e-commerce ou negócio ou fenômenos que tenha alta variabilidade de produtos e frequência de compra.
